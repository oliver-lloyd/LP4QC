```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
attributes = pd.read_csv('node_attributes.csv')
attributes['prefix'] = ['-'.join(string.split('-')[:2]) for string in attributes.id]
attributes.head()
```

# Prefix distribution

```python
#plt.figure(figsize=(8, 12))
order = sorted(attributes.prefix.unique())
sns.countplot(y='prefix', data=attributes,order=order)
plt.show()
```

# Consortium distributions

```python
# Consortium distribution
plt.figure(figsize=(6, 8))
sns.countplot(y=attributes.consortium)
plt.show()
```

# UK Biobank consortiums

```python
ukbs = [name.startswith('ukb') for name in attributes.id]
plt.figure(figsize=(6, 8))
sns.countplot(y=attributes.loc[ukbs].consortium, dodge=True)
plt.show()
```
